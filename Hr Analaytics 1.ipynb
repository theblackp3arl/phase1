```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('hrea.csv')
```

```python
print ('Dataset Information:')
print(data.info())
```

```python
print('\nFirst few rows of the dataset:')
print(data.head())
```

```python
attrition_summary = data['Attrition'].value_counts(normalize=True) *100
print ('\nAttrition Summary (%):')
print(attrition_summary)
```

```python
 plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=data, palette='Set2')
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='Age', data=data, palette='Set2')
plt.title('Attrition vs Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data, palette='Set2')
plt.title('Attrition vs MonthlyIncome')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='JobRole', hue='Attrition', data=data, palette='Set2')
plt.title('Attrition vs JobRole')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Department', hue='Attrition', data=data, palette='Set2')
plt.title('Attrition vs Department')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data, palette='Set2')
plt.title('Attrition vs JobSatisfaction')
plt.show()
```

```python
data_encoded = pd.get_dummies(data, drop_first=True)
```

```python
corr_matrix = data_encoded.corr()
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap ='coolwarm',vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
